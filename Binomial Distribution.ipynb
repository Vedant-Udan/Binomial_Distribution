## Binomial Distribution Graph with respective n.

```python
# importing required libraries.
import matplotlib.pyplot as plt
import scipy 
import seaborn as sns
```

```python
p = 0.7 # Probability of success.
n = [x for x in range(100)] # list of values of number of trials
binomial = [scipy.stats.binom.stats(n,p) for n in range(100)] # list of values of mean and variance of binomial function
# plotting the graphs of n vs (mean,variance of binomial function )
sns.set_style("darkgrid")
plt.plot(n, binomial)
plt.legend(["Mean","Variance"])
plt.title("Mean and Variance distribution of binomial distribution with respective n")
plt.xlabel("n")
plt.ylabel("Value")
plt.show()
```

```python
# if number of trails is fixed.
p = 0.7 # Probability of success.
n = 100 # number of trails
r = [x for x in range(101)] # list of number of success at particular n.
binomial_r = [scipy.stats.binom.pmf(r,n,p) for r in range(101)] # value of binomial function at r success.
# plotting the graph of r vs pnf_at_r_success
sns.set_style("darkgrid")
plt.plot(r,binomial_r)
plt.title("binomial distribution graph with respective r")
plt.xlabel("r_success")
plt.ylabel("pmf_at_r_success")
plt.show()
```
